## Replacing the missing values

```python
# determine the number of missing values in each DataFrame
print("---------uk_cases DataFrame---------")
print(uk_cases.isnull().sum())

print("-------uk_vaccinated DataFrame-------")
print(uk_vaccinated.isnull().sum())

print("--------global_data DataFrame--------")
print(global_data.isnull().sum())

print("----------tweets DataFrame----------")
print(tweets.isnull().sum())
```

```python
# replace missing values in 'uk_cases' with forward fill
uk_cases['Deaths'].fillna(method='ffill', inplace=True)
uk_cases['Cases'].fillna(method='ffill', inplace=True)
uk_cases['Recovered'].fillna(method='ffill', inplace=True)
uk_cases['Hospitalised'].fillna(method='ffill', inplace=True)

# check the sum of missing values again
print("---------uk_cases DataFrame---------")
print(uk_cases.isnull().sum())
```

```python
# replacement done in Excel and import the cleaned CSV file
global_data_cleaned = pd.read_csv('global_data_cleaned.csv')

# view the DataFrame of global_data_cleaned
print(global_data_cleaned.shape)
print(global_data_cleaned.dtypes)
print(global_data_cleaned.columns)
global_data_cleaned.head()
```

## Merge the DataFrames

```python
# merge the two DataFrames
uk_merge = pd.merge(uk_cases, uk_vaccinated, 
                    how='left', 
                    on=['Province/State', 'Country/Region', 'Date'])

print(uk_merge.shape)
print(uk_merge.dtypes)
print(uk_merge.columns)
uk_merge.head()
```

```python
# drop the unnecessary columns
uk_merge.drop(columns=['Lat_x',
                       'Long_x',
                       'ISO 3166-1 Alpha 3-Codes_x',
                       'Sub-region Name_x',                       
                       'Intermediate Region Code_x',
                       'Lat_y',
                       'Long_y',
                       'ISO 3166-1 Alpha 3-Codes_y',
                       'Sub-region Name_y',                       
                       'Intermediate Region Code_y'],
             inplace=True)

print(uk_merge.shape)
print(uk_merge.dtypes)
print(uk_merge.columns)
uk_merge.head()
```

```python
uk_merge.describe()
```

```python
# convert the data type of the Date column from object to DateTime
uk_merge['Date'] = pd.to_datetime(uk_merge['Date'])

print(uk_merge.dtypes)
```

### Which region has the greatest number of people who have received first dose but not second dose?

```python
# group the merged DataFrame by regions
uk_merge_province = uk_merge.groupby('Province/State').agg(sum).reset_index()

# sort to see which regions with highest 
uk_merge_province.sort_values(['Vaccinated'], ascending=False, inplace=True)

# view the grouped DataFrame
uk_merge_province
```

**Some brief thoughts:**
   1. Those who have got the first dose, they are willing to get the second dose.
   2. Since 95% of the 1st dose group have received 2nd dose. 
   3. However, there is a minimum of 12 weeks time gap between 1st dose and 2nd dose.
   4. Therefore, a difference of 4.5% across all regions makes sense.

## Assignment Activity 4: Visualise and identify initial trends

```python
# group the merged DataFrame by regions
uk_merge_province = uk_merge.groupby('Province/State').agg(sum).reset_index()

# sort to see which regions with highest 
uk_merge_province.sort_values(['Vaccinated'], ascending=False, inplace=True)

# view the grouped DataFrame
uk_merge_province
```

```python
# add new column with ratio of people received 2nd dose over 1st dose

uk_merge_province['Vaccinated %'] = \
round(uk_merge_province['Second Dose'] / uk_merge_province['First Dose']*100, 4)

uk_merge_province
```

```python
# create an empty plot
fig_vaccine, ax = plt.subplots()

# set the figure size
fig_vaccine.set_size_inches(10, 4)

# Plot vertical bar graph
fig_vaccine = sns.barplot(x="Province/State", y="Vaccinated %", data=uk_merge_province)
```
