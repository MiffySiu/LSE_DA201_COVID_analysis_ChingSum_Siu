### 2.2 Replacing the missing values

```python
# determine the number of missing values in each DataFrame
print("---------uk_cases DataFrame---------")
print(uk_cases.isnull().sum())

print("-------uk_vaccinated DataFrame-------")
print(uk_vaccinated.isnull().sum())

print("--------global_data DataFrame--------")
print(global_data.isnull().sum())
```

**Choice of replacing missing values:**

In order to choose a reasonable way to replace the missing values, investigation was done through Excel.<br>

- For uk_cases data set, there are two rows with missing values which belongs to region Bermuda on 2020/09/21 and 2020/09/22. The columns are cumulative data in time series with slightly changes (except hospitalised) in that period of time. Therefore, forward fill would be a reasonable way in this case, assuming there were no changes in that period.

- For global data set, the missing cells are the countries or regions with no provinces or states. So, they are replaced with their country names or region names. It is done by filtering the empty cells and setting them equal to the adjacent cells through Excel.

Lastly, I tried to find a better way to check 0 missing values in case of too many columns in the DataFrame.

```python
# Define function to plot moving averages
def plot_moving_average(series, window, plot_intervals=False, scale=1.96):
    
    # Create a rolling window to calculate the rolling mean using the series.rolling function
    rolling_mean = series.rolling(window=window).mean()
    
    # Declare the dimensions for the plot, plot name and plot the data consisting of the rolling mean from above. 
    plt.figure(figsize=(18,4))
    plt.title('Moving average\n window size = {}'.format(window))
    plt.plot(rolling_mean, 'g', label='Simple moving average trend')

    
    # Plot confidence intervals for smoothed values
    if plot_intervals:
        
        # Calculate the mean absolute square 
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        
        # Calculate the standard deviation using numpy's std function
        deviation = np.std(series[window:] - rolling_mean[window:])
        
        # Calculate the upper and lower bounds 
        lower_bound = rolling_mean - (mae + scale * deviation)
        upper_bound = rolling_mean + (mae + scale * deviation)
        
        # Name and style upper and lower bound lines and labels 
        plt.plot(upper_bound, 'r--', label='Upper bound / Lower bound')
        plt.plot(lower_bound, 'r--')
    
    # Plot the actual values for the entire timeframe.
    plt.plot(series[window:], label="Actual values")
    plt.legend(loc="best")
    plt.grid(True)
```

```python
# plot the moving average and the actual data
plot_moving_average(sample_ci.Hospitalised, 7)
```

```python
# plot with upper and lower bond
plot_moving_average(sample_ci.Hospitalised, 7, plot_intervals=True)
```
